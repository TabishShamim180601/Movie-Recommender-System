```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
#merging the dataframes based on title (inner join)
#since both datasets have same number of rows, inner join can safely be done
movies = movies.merge(credits,on='title')
```

Getting only the necessary columns

```python
#id is needed for movie poster in website
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.info()
```

```python
#checking for missing data

movies.isnull().sum()
```

```python
#only 3 rows have overview as missing. Deleting those rows

movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#checking for duplicate rows 
movies.duplicated().sum()
```

Formatting genres, keywords, cast and crew columns

```python
movies.iloc[0].genres
```

```python
import ast

def convert_genres_keywords(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert_genres_keywords)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert_genres_keywords)
```

```python
movies.head()
```

```python
#in cast keeping only top 3 actors

def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter=counter+1 
        else:
            break 
    
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
#in crew column, only keeping name of director

def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break 
    return L
```

```python
movies['director'] = movies['crew'].apply(get_director)
movies=movies.drop('crew',axis=1)
```

```python
movies.head()
```

```python
#converting overview column to list 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing spaces between words in order to create efficient tags
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director']=movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

Creating tags column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['director']
```

```python
movies.head()
```

```python
#keeping only id, title and tags column

movies = movies[['movie_id','title','tags']]
```

```python
movies.head()
```

```python
#converting tags to string 
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies['tags'][0]
```

```python
#converting tags to lowercase 
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
movies.head()
```

```python
#applying stemming on tags
%pip install nltk

import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
movies['tags'].apply(stem)
```

Vectorising the tags

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
#getting cosine similarity of a vector with every other vector

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
#method to recommend movie based on input movie name provided by the user
def recommend(movie):
    #getting index of movie 
    movie_index = movies[movies['title'].str.lower() == movie.lower()].index[0]
    #getting similarity score with each movie
    similarities = similarity[movie_index]
    #getting 5 most similar movies 
    movies_list = sorted(list(enumerate(similarities)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
        
```

```python
recommend('batman begins')
```
